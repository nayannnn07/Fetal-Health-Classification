` 5. Split into independent and dependent features(X, y) `

* X-independent(all except fetal_health)
* y-dependent(fetal_health)

```python
X = data.drop('fetal_health', axis=1)
y = data['fetal_health']
```

`6. Train-Test Split `

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, train_size = 0.20, random_state = 33
    )
```

` 7. Logistic Regression `

```python
from sklearn.linear_model import LogisticRegression
classify_model = LogisticRegression(max_iter = 100)
classify_model.fit(X_train, y_train)
```

` 8. Classify`

```python
preds = classify_model.predict(X_test)
preds
```

` 9. Prediction`

```python
comp = pd.DataFrame({'y_test':y_test, 'y_preds': preds})
comp
```

` 10. Model Evaluation`

```python
from sklearn.metrics import accuracy_score, classification_report
acc_lr = accuracy_score(y_test, preds)
print('Accuracy score: ',acc_lr)
print('Classification Report:')
print(classification_report(y_test, preds))
```

` CONFUSION MATRIX FOR LOGISTIC REGRESSION`

```python
from sklearn.metrics import confusion_matrix
conf_matrix = confusion_matrix(y_test, preds)
print("Confusion Matrix:")
print(conf_matrix)
ax= plt.subplot()
sns.heatmap(conf_matrix, annot=True,ax= ax, cmap="PuBuGn")
ax.set_xlabel("Predicted Labels")
ax.set_ylabel("True labels")
ax.set_title("Confusion Matrix for Logistic Regression")
ax.xaxis.set_ticklabels(["Normal", "Suspect", "Pathological"])
```

` DECISION TREE`

```python
from sklearn.tree import DecisionTreeClassifier
dt_classifier = DecisionTreeClassifier()
dt_classifier.fit(X_train, y_train)
dt_prediction = dt_classifier.predict(X_test)
acc_dt = accuracy_score(y_test, dt_prediction)
print('Accuracy score: ',acc_dt)
print('Classification Report:')
print(classification_report(y_test, dt_prediction))
```

` CONFUSION MATRIX FOR DECISION TREE`

```python
conf_matrix = confusion_matrix(y_test, dt_prediction)
print("Confusion Matrix:")
print(conf_matrix)
ax= plt.subplot()
sns.heatmap(conf_matrix, annot=True, cmap="PuBuGn", ax= ax)
ax.set_xlabel("Predicted Labels")
ax.set_ylabel("True labels") 
ax.set_title("Confusion Matrix for Decision Tree")
ax.xaxis.set_ticklabels(["Normal", "Suspect", "Pathological"])
```

` K-NEAREST NEIGHBOR(K-NN)`

```python
from sklearn.neighbors import KNeighborsClassifier  
knn = KNeighborsClassifier(n_neighbors=3) 
knn_mod =  knn.fit(X_train, y_train)
pred_knn = knn_mod.predict(X_test)
acc_knn = accuracy_score(y_test, pred_knn)
print('Accuracy score: ',acc_knn)
print("Classification Report:")
print(classification_report(y_test, pred_knn))
```

` CONFUSION MATRIX FOR KNN`

```python
conf_matrix = confusion_matrix(y_test, pred_knn)
print("Confusion Matrix:")
print(conf_matrix)
ax= plt.subplot()
sns.heatmap(conf_matrix, annot=True, cmap="PuBuGn", ax= ax)
ax.set_xlabel("Predicted Labels")
ax.set_ylabel("True labels") 
ax.set_title("Confusion Matrix for K-NN")
ax.xaxis.set_ticklabels(["Normal", "Suspect", "Pathological"])
```

`RANDOM FOREST`

```python
from sklearn.ensemble import RandomForestClassifier
random_forest = RandomForestClassifier()
random_forest_mod = random_forest.fit(X_train, y_train)
pred_random_forest = random_forest_mod.predict(X_test)
acc_rf = accuracy_score(y_test, pred_random_forest)
print('Accuracy score: ',acc_rf)
print("Classification Report:")
print(classification_report(y_test, pred_random_forest))
```

` CONFUSION MATRIX FOR RANDOM FOREST`

```python
conf_matrix = confusion_matrix(y_test, pred_random_forest)
print("Confusion Matrix:")
print(conf_matrix)
ax= plt.subplot()
sns.heatmap(conf_matrix, annot=True, cmap="PuBuGn", ax= ax)
ax.set_xlabel("Predicted Labels")
ax.set_ylabel("True labels") 
ax.set_title("Confusion Matrix for Random Forest")
ax.xaxis.set_ticklabels(["Normal", "Suspect", "Pathological"])
```

`ACCURACY COMPARISON`

```python
models = ['Logistic Regression','Decision Tree','K-NN','Random Forest']
acc_scores = [acc_lr,acc_dt,acc_knn,acc_rf]
results = pd.DataFrame({"Model": ["Logistic Regression","Decision Tree", "K-NN","Random Forest"], "Score":[acc_lr, acc_dt, acc_knn, acc_rf]})
result_df = results.sort_values(by="Score", ascending=False)
result_df = result_df.set_index("Score")
result_df.head(5)
```

` COMPARISONS BETWEEN THE ALGORITHMS`

```python
plt.rcParams['figure.figsize'] = (10,7)
plt.bar(models, acc_scores, color=['lightblue', 'pink', 'lightgreen', 'cyan'])
plt.xlabel("Classification Algorithms")
plt.ylabel("Accuracy Scores")
plt.title("Model Comparison based on Accuracy")
plt.show()
```
